```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
# Load the Sberbank Russian Housing Market dataset
data = pd.read_csv("./train.csv")
# Scatter plot of full_sq and price_doc
sns.scatterplot(x="full_sq", y="price_doc", data=data)
plt.show()

# Histogram of life_sq
sns.histplot(data["life_sq"])
plt.show()

# Box plot of floor and price_doc
sns.boxplot(x="floor", y="price_doc", data=data)
plt.show()
```

```python
data2 = data[data['full_sq'] < 250]
sns.scatterplot(x="full_sq", y="price_doc", data=data2)
plt.show()
```

```python
#  a heatmap to visualize the correlation matrix
sns.heatmap(data[["full_sq", "life_sq", "floor", "price_doc"]].corr(), annot=True)
plt.show()

#  a pairplot to visualize the relationships between the attributes
sns.pairplot(data[["full_sq", "life_sq", "floor", "price_doc"]])
plt.show()
```

```python
# Use Mean Encoding to encode the 'floor' attribute
mean_encoding = data.groupby("floor")["price_doc"].mean().to_dict()
data["floor_mean_encoded"] = data["floor"].map(mean_encoding)
```

```python
# Fill null values in 'life_sq' with median value
median = data['life_sq'].median()
data['life_sq'].fillna(median, inplace=True)
```

```python
# Fill null values in 'life_sq' using group mean
data["life_sq"].fillna(data.groupby("num_room")["life_sq"].transform("mean"), inplace=True)
```
